```python
import copy

small = False

if small == True:
    with open('input_small.txt', 'r') as file:
        # Read the entire file and split into lines
        lines = file.read().splitlines()
else:
    with open('input.txt', 'r') as file:
        # Read the entire file and split into lines
        lines = file.read().splitlines()

matrix = [list(line) for line in lines]
matrix2 = matrix.copy()
```

```python
def valid_pos(matrix,r,c,dir):
    if (0 <= r+dir[0] <= (len(matrix) -1)) & (0 <= c+dir[1] <= (len(matrix[0]) -1)):
        return True
    else:
        return False

def turn_90(dir):
    if dir == [-1,0]:
        dir = [0,1]
    elif dir == [0,1]:
        dir = [1,0]
    elif dir == [1,0]:
        dir = [0,-1]
    elif dir == [0,-1]:
        dir = [-1,0]
    return dir
```

```python
#### START POSITION ####
for r in range(len(matrix)-1):
    for c in range(len(matrix[0])-1):
        if matrix[r][c] == '^':
            print(f'found start pos at {r,c}')
            start_pos = [r,c]
matrix[start_pos[0]][start_pos[1]] = '.'
```

```python
dir = [-1,0]
visited = [start_pos.copy()]
cr,cc = start_pos.copy()
leaving = False
while not leaving:
    if valid_pos(matrix,cr,cc,dir):
        # print(f'valid position at {cr+dir[0],cc+dir[1]}')
        if (matrix[cr+dir[0]][cc+dir[1]] == '.'):
            cr,cc = cr+dir[0],cc+dir[1]
            if [cr,cc] not in visited:
                visited.append([cr,cc])
            continue
        elif (matrix[cr+dir[0]][cc+dir[1]] == '#'):
            dir = turn_90(dir)
    else:
        leaving = True


print(f'Answer 1: {len(visited)}')
```

```python
def translate_dir(dir):
    if dir == [-1,0]:
        sign = 'u'
    elif dir == [0,1]:
        sign = 'r'
    elif dir == [1,0]:
        sign = 'd'
    elif dir == [0,-1]:
        sign = 'l'
    else:
        sign = ''
    return sign

def check_direction(matrix,r,c,dir):
    sign = translate_dir(dir)
    if(matrix[r+dir[0]][c+dir[1]] == sign):
        return True

def state_key(cr, cc, dir):
    return (cr, cc, tuple(dir))
```

```python
#### START POSITION ####
for r in range(len(matrix2)-1):
    for c in range(len(matrix2[0])-1):
        if matrix2[r][c] == '^':
            print(f'found start pos at {r,c}')
            start_pos = [r,c]
matrix2[start_pos[0]][start_pos[1]] = 'u'

matrix_bu = matrix2.copy()
```

```python
dir = [-1, 0]  # Starting direction
cr, cc = start_pos.copy()
loop_counter = 0

for row in range(len(matrix_bu)):

    for col in range(len(matrix_bu[0])):
        dir = [-1,0]
        cr,cc = start_pos.copy()
        matrix = copy.deepcopy(matrix_bu)
        matrix[row][col] = '#'  # Set obstacle
        leaving = False
        visited = set()  
        while not leaving:
            if valid_pos(matrix, cr, cc, dir):
                next_r, next_c = cr + dir[0], cc + dir[1]
                if matrix[next_r][next_c] in ['.', 'u', 'r', 'l', 'd']:
                    cr, cc = next_r, next_c
                    state = state_key(cr, cc, dir)
                    if state in visited:
                        loop_counter += 1
                        leaving = True 
                    else:
                        visited.add(state)
                        matrix[cr][cc] = translate_dir(dir)
                elif matrix[next_r][next_c] == '#':
                    dir = turn_90(dir) 
            else:
                leaving = True 

print(f'Answer 2: {loop_counter}')

```
