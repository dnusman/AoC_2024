```python
small = False

if small == True:
    with open('input_small.txt', 'r') as file:
        lines = file.read().split(' ')
else:
    with open('input.txt', 'r') as file:
        lines = file.read().split(' ')

stones = lines
```

```python
curr_stones = stones.copy()
stone_total = [len(curr_stones)]
for blink in range(35):

    new_stones = []
    for stone in range(len(curr_stones)):
        if curr_stones[stone] == '0':
            new_stones += ['1']
        elif len(curr_stones[stone]) % 2 == 0:
            left = str(int(curr_stones[stone][:len(curr_stones[stone])//2]))
            right = str(int(curr_stones[stone][len(curr_stones[stone])//2:]))
            new_stones += [left,right]
        else:
            new_stones.append(str(int(curr_stones[stone]) * 2024))
    stone_total.append(len(new_stones))
    curr_stones =new_stones.copy()  

print(len(curr_stones))
```

```python
import matplotlib.pyplot as plt

# Plot the values with index on x-axis
plt.plot(range(len(stone_total)), stone_total, marker='o')
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Plot of Values vs Index')
plt.grid(True)
plt.show()
```
